```python
string = "A"*37
string_bits = string.encode("utf-8")

string_enc = "df765e8143ba5aa396b36b8ca894492640e089b71d92f6b33fcb9d64c702b8066606df0a37"
cipher_bits = bytes.fromhex(string_enc)

keystream = bytes(c^p for c,p in zip(cipher_bits,string_bits))

vault_hex = "c97e4bbb34cf28d7e1c51cf5deb33c5e6496fa906de2d4cb47efba16b773c8244624fd7f0b"
vault_bits = bytes.fromhex(vault_hex)
vault_plain = bytes(c^k for c,k in zip(vault_bits,keystream))

print(vault_plain.decode())
```
